```python
os.listdir()
```

```python
path = '../celestini-audio-dataset/data/'
```

```python
with open('rk.txt') as f:
    f_list = list(f)
f_list
```

```python
changes = {}
for i in range(len(f_list)):
    l = f_list[i].strip().split(':')
    changes[l[0]] = l[1]
changes
```

```python
for key in changes.keys():
    os.rename(path+key, path+changes[key])
```
